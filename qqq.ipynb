```python
from obspy.taup import TauPyModel
from obspy.taup import plot_travel_times
import matplotlib.pyplot as plt
model = TauPyModel(model="iasp91")
```

```python

arrivals = model.get_travel_times(source_depth_in_km=19,
                                  distance_in_degree=1,phase_list=["P",'p', "S",'s', "4kmps",'3.8kmps'])
fig, ax = plt.subplots(figsize=(9, 9))
ax = plot_travel_times(source_depth=10, phase_list=["P",'p', "S",'s', "4kmps"],
                       ax=ax, fig=fig, verbose=True)
ax.set_xlim(0,5)
ax.set_ylim(0,10)
# fig.savefig('qwqw.pdf')
```
